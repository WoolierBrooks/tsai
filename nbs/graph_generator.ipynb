```python
import os
import numpy as np
import pandas as pd

# Carregar dados de pacientes com diagnóstico de demência positivo
folder_dementia = r"C:\Users\Lenovo\Desktop\IC\[99] Database Final (16_4296)\cookie_d"
data_dementia = []
for npy_file in os.listdir(folder_dementia):
    if npy_file.endswith(".npy"):
        data = np.load(os.path.join(folder_dementia, npy_file))
        data_dementia.append(data)

# Carregar dados de pacientes con diagnóstico de demência control (controle)
folder_control = r"C:\Users\Lenovo\Desktop\IC\[99] Database Final (16_4296)\cookie_c/"
data_control = []
for npy_file in os.listdir(folder_control):
    if npy_file.endswith(".npy"):
        data = np.load(os.path.join(folder_control, npy_file))
        data_control.append(data)

# Combine dementia and control data
X = np.concatenate([data_dementia, data_control], axis=0)
y = np.concatenate([np.ones(len(data_dementia)), np.zeros(len(data_control))])
```

```python
from sklearn import preprocessing
X_nor = preprocessing.normalize(X)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Carregar a paleta de cores "hls"
palette = sns.color_palette("Set2")

# Taxa de amostragem
sample_rate = 16

# Número total de pontos
total_points = 4296

# Criar o array de valores de tempo
time_values = np.arange(0, total_points) / sample_rate

# Criar o gráfico
plt.figure(figsize=(10, 6), dpi=300)

for i in range(306):
    plt.plot(time_values, X_nor[i], color=palette[i % 8], alpha=0.7)

plt.title('Espectrograma de Voz de 306 Pacientes com Demência no Teste Cookie')
plt.xlabel('Tempo (segundos)')
plt.ylabel('Amplitude do Sinal de Áudio')
plt.savefig('Dementia.png', dpi=300, bbox_inches='tight')
plt.show()
```
